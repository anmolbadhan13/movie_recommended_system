```python
import pandas as pd
import numpy as np
```

```python
import ast # convert string into list
def convert(obj):
  L=[]
  for i in ast.literal_eval(obj):
    L.append(i['name'])
  return L
```

```python
movies['genres']=movies['genres'].apply(convert)
```

```python
movies.head()
```

```python
movies['keywords']=movies['keywords'].apply(convert)
```

```python
movies.head()
```

```python
movies['cast'][0]
```

```python
def convert3(obj):
  L=[]
  counter=0
  for i in ast.literal_eval(obj):
    if counter!=3:
         L.append(i['name'])
         counter+=1
    else:
        break
  return L
```

```python
movies['cast']=movies['cast'].apply(convert3)
```

```python
movies.head()
```

```python
movies['crew'][0]
```

```python
def fetch_director(obj):
  L=[]
  for i in ast.literal_eval(obj):
    if i['job']=='Director':
         L.append(i['name'])
         break
  return L
```

```python
movies['crew']=movies['crew'].apply(fetch_director)
```

```python
import nltk
from nltk.stem.porter import PorterStemmer

def stem(text):
    ps = PorterStemmer()  # Create an instance of PorterStemmer
    y = []
    for i in text.split():
        y.append(ps.stem(i))
    return " ".join(y)

stem('loved loving')
```
